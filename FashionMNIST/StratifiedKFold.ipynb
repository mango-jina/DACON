### Library

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import StratifiedKFold

from collections import Counter
```

### Train & Validation Set 나누기


```python
train_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DACON/Fashion_MNIST/data/train.csv')
```

```python
# train_data 갯수만큼 1을 담는 넘파이 배열 X, label을 넘파이 배열로 바꾼 y
X = np.ones((len(train_data),1)) 
y = np.array(train_data['label'])

# n_splits = fold 갯수
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 나눠준 fold를 확인해보고 csv로 저장
for fold_ind, (train_idx, val_idx) in enumerate(cv.split(X,y)):  
    print(f"TRAIN {fold_ind}:", [train_idx])
    print(" ", y[train_idx])
    train_data.loc[train_idx].to_csv(f'/content/drive/MyDrive/Colab Notebooks/DACON/Fashion_MNIST/data/train{fold_ind}.csv', index = None)
    print(f" TEST {fold_ind}:", [val_idx])
    print(" ", y[val_idx])
    train_data.loc[val_idx].to_csv(f'/content/drive/MyDrive/Colab Notebooks/DACON/Fashion_MNIST/data/val{fold_ind}.csv', index = None)
```

```python
# label 별 갯수 세는 함수
def get_counts(y): 
    y_count = Counter(y)
    return [f'{y_count[i]}' for i in range(np.max(y) +1)]


# train_data의 총 label 별 데이터 개수 담아주기
counts = [get_counts(y)]
index = ['training set']

# fold 별로 클래스 비율이 어떻게 나뉘었는지 확인하기
for fold_ind, (train_idx, val_idx) in enumerate(cv.split(X,y)): 
    train_y, val_y = y[train_idx], y[val_idx]

    counts.append(get_counts(train_y)) 
    counts.append(get_counts(val_y))
    index.append(f'train - fold{fold_ind}')
    index.append(f'val - fold{fold_ind}')

# DataFrame으로 만들어 확인 
pd.DataFrame(counts, index=index, columns = [i for i in range(10)])
```
